```python
def get_features(img):

    return list(cv2.mean(img)[:-1])
```

```python
data = []
features = []
basepath='./dataset/'
for folder in os.scandir(basepath):
  if folder.is_dir():
    #print(folder.name)
    for file in os.scandir(folder.path):
            im = cv2.imread(file.path,cv2.IMREAD_COLOR)
            im = cv2.cvtColor(im,cv2.COLOR_BGRA2RGB)
            #print(im.shape)
            features = get_features(im)
            #print(type(features))
            features.append(folder.name)
            #print(folder.name)
            data.append(features)
            #print(features)


print(data)
```

```python
df = pd.DataFrame(data, columns=['R','G','B','class'])
print(df)
df.to_csv('curdata.csv',index=False)
```

```python
df = pd.read_csv('curdata.csv')

print(df)

aX = df[df['class']=='apple']
oX = df[df['class']=='orange']
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

%matplotlib inline

plt.figure(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='k')
ax = plt.axes(projection='3d')

ax.set_xlabel('R')
ax.set_ylabel('G')
ax.set_zlabel('B');

ax.scatter3D(aX['R'], aX['G'], aX['B'],color='y')
ax.scatter3D(oX['R'], oX['G'], oX['B'],color='g')
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
y=le.fit(y).transform(y)

print(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.neural_network import MLPClassifier
#from sklearn.svm import SVC

classifier = MLPClassifier(random_state=1,hidden_layer_sizes=(100, 50),max_iter=1000)
#classifier = SVC()

classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

print(y_pred)
print(y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from PIL import Image
import matplotlib.pyplot as plt

image=input("Enter image path : ")
im=np.array(Image.open(image))
features=get_features(im)
yp=classifier.predict([features])
yp=le.inverse_transform(yp)
print('It is a '+yp[0])
plt.axis("off")
plt.imshow(im)
plt.show()
```
