###plot distribution of energy across different frequency bands and mfccs


```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load audio file
audio_file = 'horse2.wav'
y, sr = librosa.load(audio_file)

# Compute Mel spectrogram
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)

# Convert to decibels
S_db = librosa.power_to_db(S, ref=np.max)

# Plot Mel spectrogram
plt.figure(figsize=(10, 4))
librosa.display.specshow(S_db, x_axis='time', y_axis='mel', sr=sr, fmax=8000)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram of Horse-2')
plt.tight_layout()

# Compute mean energy across frequency bands
energy = np.mean(S, axis=1)

# Plot distribution of energy across frequency bands
plt.figure(figsize=(8, 4))
plt.bar(np.arange(len(energy)), energy)
plt.xlabel('Mel band')
plt.ylabel('Energy')
plt.title('Distribution of energy across frequency bands of Horse-2')
plt.tight_layout()

plt.show()
```
